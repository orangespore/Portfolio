```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Machine Learning : 
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, IsolationForest
from sklearn import linear_model as lm
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn import svm, metrics
from imblearn.over_sampling import SMOTENC
from collections import Counter

# Machin Learning :   
from sklearn.metrics import accuracy_score, roc_curve, auc, mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer

def read_data_preproc(path_data_full, X, Y, lst_cols_remove):
    # Data Loading
    df = pd.read_excel(path_data_full)
    df = df.drop(columns=lst_cols_remove)
    return df

#IQR   
def detect_outliers(df, n, features): 
    outlier_indices = [] 
    for col in features: 
        Q1 = np.percentile(df[col], 25) 
        Q3 = np.percentile(df[col], 75) 
        IQR = Q3 - Q1 
        outlier_step = 1.5 * IQR 
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index 
        outlier_indices.extend(outlier_list_col) 
    outlier_indices = Counter(outlier_indices) 
    return outlier_indices

def generate_data_outlier(df, cnt_cols, lst_cols):
    Outliers_to_drop = detect_outliers(df, cnt_cols, lst_cols)
    df = df.drop(index=Outliers_to_drop).reset_index(drop=True)
    return df

def viz_data_preproc(df):
    #viz 1
    print('viz 1 : correlation (by Max_Capacity) of all variables (after preprocessing)')
    #  corr 
    cancel_corr = df.corr()["Max_Capacity"]
    print(cancel_corr.abs().sort_values(ascending=False)[1:])
    print('===================================================================================')
    print('===================================================================================')      
    #viz 2
    print('viz 2 : check Skewness / Kurtosis of numeric features (after preprocessing)')
    #  Skewness Kurtosis 
    for col in df:
        print ('{:30}'.format(col), 
               'Skewness: {:05.2f}'.format(df[col].skew()),
               '    ',
               'Kurtosis: {:06.2f}'.format(df[col].kurt())
              )
    print ('\n')
    print('===================================================================================')
    print('===================================================================================')      
    #viz 3
    print('viz 2-1 : transform and check numeric features (after preprocessing)')
    #Skewness 3  Normalization
    Skewed_Features = []

    for i in Skewed_Features:
        df[i] = np.log1p(df[i])

    # log  Skewness Kurtosis 
    print("Skewness and Kurtosis of numeric features")
    for col in df:
        print ('{:30}'.format(col), 
               'Skewness: {:05.2f}'.format(df[col].skew()),
               '    ',
               'Kurtosis: {:06.2f}'.format(df[col].kurt())
              )
    print ('\n')
    print('===================================================================================')
    print('===================================================================================')      
    #viz 4
    print('viz 3 : check outlier (after preprocessing)')
    plt.figure(figsize = (15, 15))
    df.boxplot(rot = 45)
    plt.title("Box Plot Before Outlier Elimination", fontsize=16)
    plt.show()
    print('===================================================================================')
    print('===================================================================================')      
    #viz 5
    print('viz 3-1 : check after removing outlier (after preprocessing)')
    # Outlier   outlier  Box plot 
    plt.figure(figsize = (15, 15))
    df.boxplot(rot = 45)
    plt.title("Box Plot After Outlier Elimination : X variables", fontsize=16)
    plt.show()
    print('===================================================================================')
    print('===================================================================================')  
    
def generate_XY(df):
    #   
    # y  Series  DF  
    x_df = df.drop(columns=["Max_Capacity"])
    y = df["Max_Capacity"]
    y_df = pd.DataFrame(data = y, dtype = np.float64)
    return x_df, y_df

def preprocess_XY(x_df, y_df):
    #  
    #   shape error   x y scaler  
    x_scaler = MinMaxScaler(feature_range = (0,1))
    y_scaler = MinMaxScaler(feature_range = (0,1))
    x_df_cln = x_scaler.fit_transform(x_df)
    y_df_cln = y_scaler.fit_transform(y_df)

    #     NumPy ndarray    DataFrame 
    # rangeindex     df_cln index  
    x_df = pd.DataFrame(data = x_df_cln, index = x_df.index, columns = x_df.columns, dtype = np.float64)
    y_df = pd.DataFrame(data = y_df_cln, index = y_df.index, columns = y_df.columns, dtype = np.float64)
    
    #   Preprocessing
    #    preprocessing       
    #  missing value            
    num_transformer = SimpleImputer(strategy="constant", fill_value = 0)
    # Bundle preprocessing for numerical and categorical features:
    preprocessor = ColumnTransformer(transformers=[("num", num_transformer, X)])
    #    (df )
    X_train, X_test, Y_train, Y_test = train_test_split(x_df, y_df.values.ravel(), test_size=.2)
    
    return X_train, X_test, Y_train, Y_test, y_scaler

def modelling(X_train, X_test, Y_train, Y_test, params, model, y_scaler):
    # Random Forest Training  Testing
    grid = GridSearchCV(model, params, cv=5, n_jobs = -1, scoring="neg_mean_squared_error", return_train_score = True)
    grid = grid.fit(X_train, Y_train)
    model = grid.best_estimator_
    Y_test_pred = model.predict(X_test)

    print("model_name : ", model.__repr__())
    print("best params : ", grid.best_params_)  
    
    # MAE     MinMaxScaler 
    Y_test_pred = Y_test_pred.reshape(Y_test_pred.size, 1)
    unnormalized_Y_test_pred = y_scaler.inverse_transform(Y_test_pred)

    Y_test = Y_test.reshape(Y_test.size, 1)
    unnormalized_Y_test = y_scaler.inverse_transform(Y_test)
    
    return unnormalized_Y_test, unnormalized_Y_test_pred, model

def evaluation(unnormalized_Y_test, unnormalized_Y_test_pred):
    Test_mae = MAE(unnormalized_Y_test, unnormalized_Y_test_pred)
    Test_mse = MSE(unnormalized_Y_test, unnormalized_Y_test_pred)
    Test_rmse = RMSE(unnormalized_Y_test, unnormalized_Y_test_pred)
    Test_mape = MAPE(unnormalized_Y_test, unnormalized_Y_test_pred)
    Test_mpe = MPE(unnormalized_Y_test, unnormalized_Y_test_pred)
    Test_r2 = r2_score(unnormalized_Y_test, unnormalized_Y_test_pred)

    # Y_test 
    unnormalized_Y_test_Avg = np.average(unnormalized_Y_test)

    print(f"Model               : {model}")
    print('Target Variable Avg : %.2f'   %(unnormalized_Y_test_Avg))
    print('R2                  : %.4f' %(Test_r2))  
    print('MAE                 : %.2f' %(Test_mae))
    print('MSE                 : %.2f' %(Test_mse))
    print('RMSE                : %.2f' %(Test_rmse))
    print('MAPE                : %.2f' %(Test_mape))
    print('MPE                 : %.2f' %(Test_mpe))
    
def viz_result(model, df, figsize=(15, 18)):
    sns.set_style('whitegrid')
    
    feature_importance = model.feature_importances_
    feature_importance_rel = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    sorted_idx_rel = np.argsort(feature_importance_rel)
    ypos = np.arange(sorted_idx_rel.shape[0]) + .5
    
    n=0
    print("Feature Importance of ", model.__repr__())
    for index in feature_importance:   
        print ('{:30}'.format(df.columns[n]), 
               'importance: {:05.4f}'.format(index)
              )
        n = n+1
        
    plt.figure(figsize=figsize)
    plt.barh(ypos, feature_importance[sorted_idx_rel], align='center', color = "Green")
    plt.yticks(ypos, df.columns[sorted_idx_rel])
    plt.xlabel('Relative importance')
    plt.title('Feature importance')
    plt.show()
    
# Feature Importance
def plot_feature_importance(model, df, figsize=(15, 18)):
    sns.set_style('whitegrid')
    
    feature_importance = model.feature_importances_
    feature_importance_rel = 100.0 * (feature_importance / feature_importance.max())
    sorted_idx = np.argsort(feature_importance)
    sorted_idx_rel = np.argsort(feature_importance_rel)
    ypos = np.arange(sorted_idx_rel.shape[0]) + .5
    
    n=0
    print("Feature Importance of ",model.__repr__())
    for index in feature_importance:   
        print ('{:30}'.format(df.columns[n]), 
               'importance: {:05.4f}'.format(index)
              )
        n = n+1
        
    plt.figure(figsize=figsize)
    plt.barh(ypos, feature_importance[sorted_idx_rel], align='center', color = "Green")
    plt.yticks(ypos, df.columns[sorted_idx_rel])
    plt.xlabel('Relative importance')
    plt.title('Feature importance')
    plt.show()

#     
def MAE(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred)))
def MSE(y_true, y_pred):
    return np.mean(np.square((y_true - y_pred)))
def RMSE(y_true, y_pred):
    return np.sqrt(MSE(y_true, y_pred))
def MAPE(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
def MPE(y_true, y_pred): 
    return np.mean((y_true - y_pred) / y_true) * 100

#     X, Y
X = 40
Y = 60
path_data_full = "/home/heemok/km/km_capacity_prediction/Full_Data_X{}Y{}.xlsx".format(X,Y)
lst_cols_remove = ["P_VKOSPI_T1","P_VKOSPI_T2","P_VKOSPI_T3","P_VKOSPI_T4","P_VKOSPI_T5","P_VKOSPI_T6","P_VKOSPI_T7",
                      "P_VKOSPI_T8","P_VKOSPI_T9","P_VKOSPI_T10","P_VKOSPI_T11","P_VKOSPI_T12","P_VKOSPI_T13","P_VKOSPI_T14",
                      "P_VKOSPI_T15","P_VKOSPI_T16","P_VKOSPI_T17","P_VKOSPI_T18","P_VKOSPI_T19","P_VKOSPI_T20", 
                      "P_VKOSPI_T21","P_VKOSPI_T22","P_VKOSPI_T23","P_VKOSPI_T24","P_VKOSPI_T25","P_VKOSPI_T26","P_VKOSPI_T27",
                      "P_VKOSPI_T28","P_VKOSPI_T29","P_VKOSPI_T30","P_VKOSPI_T31","P_VKOSPI_T32","P_VKOSPI_T33","P_VKOSPI_T34",
                      "P_VKOSPI_T35","P_VKOSPI_T36","P_VKOSPI_T37","P_VKOSPI_T38","P_VKOSPI_T39","P_VKOSPI_T40","P_VKOSPI_T41",
                      "P_VKOSPI_T42","P_VKOSPI_T43","P_VKOSPI_T44","P_VKOSPI_T45","P_VKOSPI_T46","P_VKOSPI_T47",
                      "P_VKOSPI_T48","P_VKOSPI_T49","P_VKOSPI_T50","P_VKOSPI_T51","P_VKOSPI_T52","P_VKOSPI_T53","P_VKOSPI_T54",
                      "P_VKOSPI_T55","P_VKOSPI_T56","P_VKOSPI_T57","P_VKOSPI_T58","P_VKOSPI_T59","P_VKOSPI_T60"]

lst_params = {"bootstrap":[False], "n_estimators":[3, 10], "max_features":[4, 8, 12, 16, 20]}
model = RandomForestRegressor()

df = read_data_preproc(path_data_full, X, Y, lst_cols_remove)
df = generate_data_outlier(df, 2, df.columns)
viz_data_preproc(df)
df_X, df_Y = generate_XY(df)
X_train, X_test, Y_train, Y_test, scaler = preprocess_XY(df_X, df_Y)
lst_y_test, lst_y_pred, model = modelling(X_train, X_test, Y_train, Y_test, lst_params, model, scaler)
evaluation(lst_y_test, lst_y_pred)
viz_result(model, df_X)
```
