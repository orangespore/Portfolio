```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

def load_data(path):
    df = pd.read_excel(path)
    df['Date']= pd.to_datetime(df['Date'], format='%Y%m%d')
    return df

def save_data_corr(df, path_corr):
    df_corr = df.corr(method = 'pearson').to_excel(path_corr)

def viz_data(df, Selected_Features):
    #viz 1
    print('viz 1 : histogram of all variables (before preprocessing)')
    df.hist(bins=50, figsize=(20,15))
    plt.show()
    print('===================================================================================')
    print('===================================================================================')
    #viz 2
    print('viz 2 : heatmap(correlation) of all variables (before preprocessing)')
    plt.figure(figsize = (20, 20))
    sns.heatmap(data = df.corr(), annot = True, fmt = '.2f', linewidths = .5, cmap = 'Blues')
    plt.show()
    print('===================================================================================')
    print('===================================================================================')
    #viz 3
    print('viz 3 : heatmap(correlation by N_Order) of all variables (before preprocessing)')
    cancel_corr = df.corr()["N_Order"]
    print(cancel_corr.abs().sort_values(ascending=False)[1:])
    print('===================================================================================')
    print('===================================================================================')
    #viz 4
    print('viz 4 : histogram(correlation by N_Order) of Selected variables (before preprocessing)')
    df_sel = df[Selected_Features]
    Sel_corr = df_sel.corr()["N_Order"]
    print(Sel_corr.abs().sort_values(ascending=False)[1:])
    df_sel.hist(bins=50, figsize=(20,15))
    plt.show()
    print('===================================================================================')
    print('===================================================================================')




def preprocess_data_lag(df, Selected_Features, X, Y):
    # X에 따른 동적 lag변수 컬럼명 생성을 위하여 str concatenation
    dynamic_feature_str =""
    for i in Selected_Features:
        j = 0
        #dynamic_feature_str = dynamic_feature_str + " {}".format(i)
        while j < X:
            dynamic_feature_str = dynamic_feature_str + " {}_T{}".format(i, j+1)
            j=j+1

    #str타입을 공백 문자로 구분하여 list로 치환   
    dynamic_feature_list = dynamic_feature_str.split()

    #동적생성된 리스트를 컬럼명으로 가지는 dataframe 선언
    #컬럼들은 모두 정수형 또는 실수형으로, float64로 일괄 적용
    df_sel = df[Selected_Features]
    df_lag = pd.DataFrame(index = df_sel.index, columns = dynamic_feature_list, dtype = np.float64)
    
    # 각 lag 변수에 값 채우기
    for i in Selected_Features:
        j = 0
        from_feature =i
        while j < X:
            to_feature = from_feature + "_T{}".format(j+1)
            df_lag[to_feature] = df_sel[from_feature].shift(j+1)
            j = j + 1
    #lag변수가 NA인 행 제거(X개수만큼 제거)
    df_lag = df_lag.dropna(axis=0)
    
    return df_lag
    
def generate_target(df_lag, df, Selected_Features):
    #Y값을 이용한 타겟변수 생성
    df_sel = df[Selected_Features]
    #Y기간도안 최대치 용량 실제치를 담을 dataframe 선언
    df_max = pd.DataFrame(index = df_lag.index, columns = ['Max_Capacity'], dtype = np.float64)

    # 모든 row에 대해서 향후 Y일의 N_Order의 최대치를 구하여 타겟변수 값 생성
    for i in df_lag.index:
        if i < max(df_lag.index) - Y+1: # 인덱싱 에러를 회피하기 위하여 마지막 Y개 튜플들은 계산하지 않아야 함
            df_max.loc[i,'Max_Capacity'] = df_sel.iloc[i,0]
            j = 0
            while j < Y:
                if df_max.loc[i,'Max_Capacity'] < df_sel.iloc[i+j+1,0]:
                    df_max.loc[i,'Max_Capacity'] = df_sel.iloc[i+j+1,0]
                #print(df_max.loc[df_max.index[i],'Max_Capacity'])   
                j = j + 1
    return df_max

def generate_dataset(df1, df2, path_data_full):
    #원래 독립변수 dataframe인 df_lag와 타겟변수 dataframe인 df_max 결합
    df_lag = df1.join(df2)

    #tail 튜플들의 경우, Y개수만큼 NA인 행 제거(Y개수만큼 제거)
    df_lag = df_lag.dropna(axis=0)
    
    #excel export
    df_lag.to_excel(path_data_full, index = False) 
    
    return df_lag
    
def viz_data_2(df_lag, df, Stat_Features):
    #viz 1
    print('viz 1 : describe of P_VKOSPI_T1 (after preprocessing)')
    df_lag["P_VKOSPI_T1"].describe()
    print('===================================================================================')
    print('===================================================================================')
    #viz 2
    print('viz 2 : describe of Stat_Features (after preprocessing)')
    df_stat = df_lag[Stat_Features]
    df_stat = pd.concat([df_stat, df['Date']], axis = 1, join = 'inner', verify_integrity=True)
    df_stat.reset_index(drop=True)
    df_stat.describe()
    print('===================================================================================')
    print('===================================================================================')
    #viz 3
    print('viz 3 : plot Comparison of Major Variables (after preprocessing)')
    fig, ax1 = plt.subplots(figsize = (12, 8))
    ax2 = ax1.twinx()
    Order_line = ax1.plot(df_stat['Date'], df_stat['N_Order_T1'], color='b', linestyle='-', label = "Number of Orders")
    Vkospi_line = ax2.plot(df_stat['Date'], df_stat['P_VKOSPI_T1'], color='r', linestyle='-', label = "VKOSPI")
    frequency = 90 # x_ticks 표시 빈도수
    ax1.set_xticks(df_stat['Date'].dt.date[::frequency])
    ax1.set_xticklabels(df_stat['Date'].dt.date[::frequency], rotation = 30)
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Number[Order]')
    ax2.set_ylabel('Points[VKOSPI]')
    plt.title('Comparison of Major Variables', fontsize=20) 
    # 그래프 limit 설정
    ax1.set_ylim(0, 60000000)
    ax2.set_ylim(0, 70.0)
    # plot legends
    lines = Order_line + Vkospi_line
    labels = [l.get_label() for l in lines]
    plt.legend(lines, labels, loc = 2)
    plt.grid(False)
    plt.show()
    print('===================================================================================')
    print('===================================================================================')    
    #viz 4
    print('viz 3 : plot Comparison of Capacity Efficiency (after preprocessing)')
    df_cap = pd.DataFrame(index = df_stat.index, columns = ['Cap'], dtype = np.float64)
    cap = pd.Series(df_stat['Max_Capacity'])

    # 가장 큰 값으로 모두 채우기
    for i in df_cap.index:
            df_cap.loc[i,'Cap'] = cap.max()
            i = i + 1
    #최대치 경신한 일자 찾기
    for i in df_stat.index:
            if df_stat.loc[i,'N_Order_T1'] == cap.max() :
                print(df_stat.loc[i,'Date'])
            i = i + 1

    fig, ax1 = plt.subplots(figsize = (12, 8))

    #Plot 설정
    Order_line = ax1.plot(df_stat['Date'], df_stat['N_Order_T1'], color='b', linestyle='-', label = "Number of Orders")
    Capacity_line = ax1.plot(df_stat['Date'], df_stat['Max_Capacity'], color='g', linestyle='-', label = "Predicted_Capacity")
    Legacy_capacity_line = ax1.plot(df_stat['Date'], df_cap['Cap'], color='r', linestyle='-', label = "Legacy_Capacity") 

    # 날짜에 맞춰서 x_ticks 설정
    ax1.set_xticks(df_stat['Date'].dt.date[::frequency])
    ax1.set_xticklabels(df_stat['Date'].dt.date[::frequency], rotation = 30)
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Number[Order, Capacity]')
    plt.title('Comparison of Capacity Efficiency', fontsize=20) 

    # 그래프 limit 설정
    ax1.set_ylim(0, 60000000)

    # plot legends
    lines = Order_line + Capacity_line + Legacy_capacity_line
    labels = [l.get_label() for l in lines]
    plt.legend(lines, labels, loc = 4)
    plt.grid(False)
    plt.show()
    print('===================================================================================')
    print('===================================================================================')    

X, Y = 40, 60  # 용량예측을 위한 과거 관측기간 X, 최대 용량을 계산할 향후 기간 Y
path_data = '/home/heemok/km/km_capacity_prediction/Raw_Data.xlsx'
path_data_corr = '/home/heemok/km/km_capacity_prediction/Raw_Data_corr.xlsx'
path_data_full = "/home/heemok/km/km_capacity_prediction/Full_Data_X{}Y{}.xlsx".format(X,Y)

lst_sel_features = ["N_Order", "N_Trade", "P_Trading_Volume_F", 
                    "N_Investor", "N_Traded_Symbol_O", "P_Margin","P_K200S","P_VKOSPI"]
Stat_Features = ["N_Order_T1","P_VKOSPI_T1","Max_Capacity"]

df = load_data(path_data)
save_data_corr(df, path_data_corr)
viz_data(df, lst_sel_features)
df_preproc = preprocess_data_lag(df, lst_sel_features, X, Y)
df_target = generate_target(df_preproc, df, lst_sel_features)
df_final = generate_dataset(df_preproc, df_target, path_data_full)
viz_data_2(df_final, df, Stat_Features)
```

```python
# Data handling
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Data Loading
read_file_path = "파일경로/Raw_Data.xlsx"
df = pd.read_excel(read_file_path)
```

```python
# Data Type Cleansing
# 변수를 날짜로 인식
# 기타 다른 변수들은 모두 타입 이상없음

df['Date']= pd.to_datetime(df['Date'], format='%Y%m%d')
```

```python
# Time Trend 캡처를 위한 Lag 변수 추가로 인해 변수가 많아져 과적합이 발생할 수 있으므로, 최소한의 Feature를 manual하게 선정

# 변수간 Correaltion 확인
# excel로 추출
df.corr(method = 'pearson').to_excel('파일경로/corr_init.xlsx')

# 속성별 히스토그램
df.hist(bins=50, figsize=(20,15))
plt.show()

# heatmap
plt.figure(figsize = (20, 20))
sns.heatmap(data = df.corr(), annot = True, fmt = '.2f', linewidths = .5, cmap = 'Blues')

# 동일유형의 변수 중에 제거하기 위한 Feature Selection을 위해 타겟변수와의 Correaltion 확인
# Correlation이 높은 동일유형의 변수들 중에서 타겟변수와 Correlation이 가장 높은 변수를 유지하고 나머지를 제거
# Time Trend의 독립변수도 아니며, 실제 예측 타겟변수(X일 간의 최대 호가건수)도 아니나, 당일의 타겟변수만으로 동일유형의 변수를
# 제거하는 것이 가능할 것으로 판단됨
cancel_corr = df.corr()["N_Order"]
print(cancel_corr.abs().sort_values(ascending=False)[1:])

# (참고용)별도 엑셀로 추출한 변수간 correlation 분석을 통해 제거해야 할 변수들, 0.9 이상으로 총 37개 변수 중 11개 제거
# 추후 0.8이상도 제거할지 검토 필요, 0.8까지 제거할 경우 약 20개 제거 
corr_features = ["P_Trading_Amount_F","P_Trading_Volume_T","N_Open_Interest_T","N_Traded_Symbol_T",
                 "N_Traded_Symbol_F","N_Symbol_T","N_Symbol_F","N_Symbol_O", "P_K200F", "P_Settlement_Amount", "P_Deposit"]

# 별도 엑셀로 분석하여 time trend를 반영할 최종 변수 선택 
Selected_Features = ["N_Order", "N_Trade", "P_Trading_Volume_F", "N_Investor", "N_Traded_Symbol_O", "P_Margin",
                    "P_K200S","P_VKOSPI"]
df_sel = df[Selected_Features]
df_sel.info()
```

```python
# 용량예측을 위한 과거 관측기간 X, 최대 용량을 계산할 향후 기간 Y
X = 40
Y = 60

# X에 따른 동적 lag변수 컬럼명 생성을 위하여 str concatenation
dynamic_feature_str =""
for i in Selected_Features:
    j = 0
    #dynamic_feature_str = dynamic_feature_str + " {}".format(i)
    while j < X:
        dynamic_feature_str = dynamic_feature_str + " {}_T{}".format(i, j+1)
        j=j+1

#str타입을 공백 문자로 구분하여 list로 치환   
dynamic_feature_list = dynamic_feature_str.split( )

#동적생성된 리스트를 컬럼명으로 가지는 dataframe 선언
#컬럼들은 모두 정수형 또는 실수형으로, float64로 일괄 적용
df_lag = pd.DataFrame(index = df_sel.index, columns = dynamic_feature_list, dtype = np.float64)

#검증
df_lag.info()
df_lag.head()
```

```python
#Y값을 이용한 타겟변수 생성

#Y기간도안 최대치 용량 실제치를 담을 dataframe 선언
df_max = pd.DataFrame(index = df_lag.index, columns = ['Max_Capacity'], dtype = np.float64)

# 모든 row에 대해서 향후 Y일의 N_Order의 최대치를 구하여 타겟변수 값 생성
for i in df_lag.index:
    if i < max(df_lag.index) - Y+1: # 인덱싱 에러를 회피하기 위하여 마지막 Y개 튜플들은 계산하지 않아야 함
        df_max.loc[i,'Max_Capacity'] = df_sel.iloc[i,0]
        j = 0
        while j < Y:
            if df_max.loc[i,'Max_Capacity'] < df_sel.iloc[i+j+1,0]:
                df_max.loc[i,'Max_Capacity'] = df_sel.iloc[i+j+1,0]
            #print(df_max.loc[df_max.index[i],'Max_Capacity'])   
            j = j + 1

#검증
df_max
```

```python
#원래 독립변수 dataframe인 df_lag와 타겟변수 dataframe인 df_max 결합
df_lag = df_lag.join(df_max)

#tail 튜플들의 경우, Y개수만큼 NA인 행 제거(Y개수만큼 제거)
df_lag = df_lag.dropna(axis=0)

#검증
df_lag.info()
df_lag.head()
```
