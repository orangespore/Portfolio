```python
import pandas as pd
import glob
import os
```

# 연도별 통계자료 만들기

```python
lst_file = glob.glob("C:/Users/admin/mok/km_snowfall_project/시나리오/*/*/*")
```

```python
lst_rcp = ['rcp26','rcp45','rcp60','rcp85']
lst_rcp_data = [file for file in lst_file if 'rcp26' in file]
```

## 통계자료 생성
    1. 시나리오별 데이터 불러오기
    2. 시나리오별 연도별/월별 자료로 정리 후 통계테이블 생성

```python
os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_연도별", exist_ok=True)

for i in lst_rcp:
    os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_연도별/{}".format(i), exist_ok=True)
    
    for year in list(set([c.split("\\")[-2] for c in lst_rcp_data])):
        os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_연도별/{}/{}".format(i,year), exist_ok=True)
```

```python
for i in lst_rcp:
    lst_rcp_data = [file for file in lst_file if i in file]
    print('-'*30)
    print(i)
    print('\n')
    lst_rcp_year = glob.glob("C:/Users/admin/mok/km_snowfall_project/시나리오/{}/*".format(i))
    lst_rcp_year = [c.split("\\")[-1] for c in lst_rcp_year]

    for year in list(set([c.split("\\")[-2] for c in lst_rcp_data])):
        print(year)
        
        lst_rcp_data_year = [c for c in lst_rcp_data if c.split("\\")[-2]==year]
        
        lst_rcp_df_year = []
        for data in lst_rcp_data_year:
            df = pd.read_csv(data, encoding='cp949')
            lst_rcp_df_year.append(pd.read_csv(data, encoding='cp949'))
            
        df_year_total = pd.concat(lst_rcp_df_year)
        df_year_total_gp = df_year_total.groupby('Name').agg(Lat=('Lat', 'max'),
                                  Long=('Long', 'max'),
                                  mean_SnowDepth=('SnowDepth', 'mean'),
                                  max_SnowDepth=('SnowDepth', 'max'),
                                  min_SnowDepth=('SnowDepth', 'min')
                                  ).reset_index()
        
        file_name = i+'_'+year+'_result.csv'
        df_year_total_gp.to_csv("C:/Users/admin/mok/km_snowfall_project/시나리오_연도별/{}/{}/{}".format(i,year,file_name), encoding='cp949')
    print('-'*30)
```

## 월별 통계자료 만들기

### 파일 구성하기

```python
os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_월별", exist_ok=True)

for i in lst_rcp:
    os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_월별/{}".format(i), exist_ok=True)
    
    for year in list(set([c.split("\\")[-2] for c in lst_rcp_data])):
        os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_월별/{}/{}".format(i,year), exist_ok=True)
        
        for month in ['10','11','12','01','02','03','04']:
            os.makedirs("C:/Users/admin/mok/km_snowfall_project/시나리오_월별/{}/{}/{}".format(i,year,month), exist_ok=True)
```

### 통계자료 생성
    1. 시나리오별 데이터 불러오기
    2. 시나리오별 연도별/월별 자료로 정리 후 통계테이블 생성

```python
for i in lst_rcp:
    lst_rcp_data = [file for file in lst_file if i in file]
    print('-'*30)
    print(i)
    print('\n')
    lst_rcp_year = glob.glob("C:/Users/admin/mok/km_snowfall_project/시나리오/{}/*".format(i))
    lst_rcp_year = [c.split("\\")[-1] for c in lst_rcp_year]

    for year in list(set([c.split("\\")[-2] for c in lst_rcp_data])):
        print(year)
        
        lst_rcp_data_year = [c for c in lst_rcp_data if c.split("\\")[-2]==year]
            
        for month in ['10','11','12','01','02','03','04']:
            
            lst_rcp_data_month = [c for c in lst_rcp_data_year if c.split('_')[-1][4:6]==month]
            
            lst_rcp_df_month = []
            for data in lst_rcp_data_month:
                #df = pd.read_csv(data, encoding='cp949')
                lst_rcp_df_month.append(pd.read_csv(data, encoding='cp949'))

            df_year_total = pd.concat(lst_rcp_df_month)
            df_year_total_gp = df_year_total.groupby('Name').agg(Lat=('Lat', 'max'),
                                      Long=('Long', 'max'),
                                      mean_SnowDepth=('SnowDepth', 'mean'),
                                      max_SnowDepth=('SnowDepth', 'max'),
                                      min_SnowDepth=('SnowDepth', 'min')
                                      ).reset_index()

            file_name = i+'_'+year+month+'_result.csv'
            df_year_total_gp.to_csv("C:/Users/admin/mok/km_snowfall_project/시나리오_월별/{}/{}/{}/{}".format(i,year,month,file_name), encoding='cp949')
        print('-'*30)
```
