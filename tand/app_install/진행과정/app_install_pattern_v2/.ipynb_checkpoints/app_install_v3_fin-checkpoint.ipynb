```python
# 1) generate total data
from collections import Counter
def gen_df_total_app_count(df_app_install, n_user_sample, n_top_app) :
    """
     - df_total_app_count : 전체 데이터의 앱 설치 빈도
     - app_list_sample :  설치 수 상위 n개 앱리스트 생성
     - n_total_user : 전체 유저의 수
    """
    # 1) df_total_app_count
    df_total_app_count = pd.DataFrame.from_dict(Counter(df_app_install['total_apps'].sum()),orient='index')
    df_total_app_count = df_total_app_count.reset_index()
    df_total_app_count = df_total_app_count.rename(columns = { 0: 'count',
                                                  'index' : 'app_name'}).\
                                            sort_values(by = 'count', ascending = False)

    # 2) total_app_list_sample by n_sample(500)
    # select only apps installed more than min_install_rate and under the max_install_rate 
    app_list_sample = df_total_app_count.head(n_top_app).app_name.values
    len(app_list_sample)
    
    
    # 3) n_total_user
    n_total_user = df_app_install['adid'].nunique()
    print('number of total user : {}'.format(n_total_user))
    
    return df_total_app_count, app_list_sample, n_total_user

df_total_app_count, app_list_sample, n_total_user = gen_df_total_app_count(df_app_install = df_app_install_sample,
                                                                           n_user_sample = n_user_sample,
                                                                           n_top_app = n_top_app)
df_total_app_count.head()
app_list_sample[:5]
n_total_user
```

```python
# 2) generate target data 
def gen_df_target_app_count(df_app_install, n_user_sample, target_app) :
    """
     - df_target_app_count : 타겟 데이터의 앱 설치 빈도
     - n_target_user : 타겟 데이터를 설치한 유저의 수
    """
    # 1) df_target_app_install
    df_target_app_install = df_app_install[df_app_install['total_apps'].apply(lambda x : True if target_app in x else False)]
    
    # 2) n_target_user
    n_target_user = df_target_app_install.adid.nunique()
    print('number of target user : {}'.format(n_target_user))
    
    # 3) df_target_app_count
    df_target_app_count = pd.DataFrame.from_dict(Counter(df_target_app_install.total_apps.sum()),
                                                 orient='index')
    df_target_app_count = df_target_app_count.reset_index()
    df_target_app_count = df_target_app_count.rename(columns = { 0: 'count',
                                                                'index' : 'app_name'}).\
                                              sort_values(by = 'count', ascending = False)

    return df_target_app_install, df_target_app_count, n_target_user
target_app = '대한항공'
df_target_app_install, df_target_app_count, n_target_user  = gen_df_target_app_count(df_app_install = df_app_install_sample,
                                                                                     n_user_sample = n_user_sample,
                                                                                     target_app = target_app)
df_target_app_install.head()
df_target_app_count.head()
n_target_user
```

```python
# 3) extract_n_top_app_data for get Total rate & intersect rate
def extract_n_top_app_data(df_total_app_count, df_target_app_count, n_top_app, n_total_user, n_target_user) :
    """
    : total & target 데이터 설치 빈도 상위 N(500개)만을 활용하여 전체 설치 확률(total_rate) & 타겟 데이터의 설치 확률(intersect_rate)계산 산출
    """
    df_total_app_count_sample = df_total_app_count.head(n_top_app)
    df_target_app_count_sample = df_target_app_count.head(n_top_app)

    df_total_app_count_sample['total_rate'] = df_total_app_count_sample['count'].apply(lambda x : x /n_total_user).values
    df_target_app_count_sample['intersect_rate'] = df_target_app_count_sample['count'].apply(lambda x : x /n_target_user).values
    
    return df_total_app_count_sample, df_target_app_count_sample
df_total_app_count_sample, df_target_app_count_sample = extract_n_top_app_data(df_total_app_count = df_total_app_count, 
                                                                               df_target_app_count = df_target_app_count,
                                                                               n_top_app = n_top_app,
                                                                               n_total_user = n_total_user, 
                                                                               n_target_user = n_target_user)
df_total_app_count_sample
df_target_app_count_sample
```

```python
# 4) merge total & target
def merge_total_target(df_total_app_count_sample, df_target_app_count_sample) :
    df_app_relation_merge_fin = pd.merge(df_total_app_count_sample[['app_name', 'total_rate']],
                                         df_target_app_count_sample[['app_name', 'intersect_rate']], 
                                         on = 'app_name')
    
    # + 중복되지 못한 앱 수
    not_interect = list(set(df_total_app_count_sample.app_name.values) - set(df_target_app_count_sample.app_name.values))
    print('중복되지 못한 앱 수(설치빈도 상위 N개로 제한함에 따른 연관도 분석 불가 앱의 수) : {}'.format(len(not_interect)))
    print(f'중복되지 못한 앱 : {not_interect[:5]}')

    return df_app_relation_merge_fin

df_app_relation_merge_fin = merge_total_target(df_total_app_count_sample, df_target_app_count_sample)
df_app_relation_merge_fin.head()
```

```python
# 5) calculate relataion
def calc_relataion(df_app_relation_merge_fin, how = 'REL', rel_weight = 0.7):
    """
    - how : ABS, REL, GEO, weight 4가지중 하나 선택
    - rel_weight
    # 1. ABS_diff
    ABS_diff = intersect_rate - total_rate

    # 2. Rel_diff
    Rel_diff = (intersect_rate / total_rate)-1

    # 3. Geo_mean
    Geo_mean = np.sqrt(ABS_diff * Rel_diff)

    # 4. weighted_Geo_mean
    weight = 0.7
    weighted_Geo_mean = np.sqrt(abs(ABS_diff)**(2*1-weight) * abs(Rel_diff)**(2*weight))
    if ABS_diff < 0 :
        weighted_Geo_mean = weighted_Geo_mean * (-1)
    """
    # select relataion way
    if how == 'ABS' :
        relation_values = df_app_relation_merge_fin['intersect_rate']  - df_app_relation_merge_fin['total_rate'] 
    elif how == 'REL' :
        relation_values = df_app_relation_merge_fin['intersect_rate'] / df_app_relation_merge_fin['total_rate'] - 1
    elif how == 'GEO' :
        relation_values = np.sqrt(df_app_relation_merge_fin['intersect_rate']  * df_app_relation_merge_fin['total_rate'] )
    elif how == 'weight' :
        abs_diff = df_app_relation_merge_fin['intersect_rate']  - df_app_relation_merge_fin['total_rate']
        rel_diff = df_app_relation_merge_fin['intersect_rate'] / df_app_relation_merge_fin['total_rate'] - 1
        relation_values = np.sqrt(abs(abs_diff)**(2*1-rel_weight) * 
                                  abs(rel_diff)**(2*rel_weight)) * abs_diff.map(lambda x: -1 if x < 0 else 1) # 음수 유지
        
    # combie result
    target_relation = str(how + '_relation')
    df_app_relation_merge_fin[target_relation] = relation_values
    
    # sorting
    df_app_relation_merge_fin = df_app_relation_merge_fin[['app_name', 'total_rate', 'intersect_rate', target_relation]].sort_values(by = target_relation, ascending = False)
    
    return df_app_relation_merge_fin
df_app_relation_merge_fin = calc_relataion(df_app_relation_merge_fin, how = 'REL', rel_weight = 0.7)
df_app_relation_merge_fin
```

# 2. relataion main 

```python
def app_install_relation_main(df_app_install, target_app, n_user_sample, n_top_app, how  = 'REL') :
    """
    : 앱설치 패턴 분석에 있어서, 유저수 및 상위 앱의 개수 설정하여 앱설치 패턴 진행
    [input]
    - df_app_install : 유저별 앱설치 데이터 
    - n_user_sample : 전체 유저중 샘플링할 유저의 수
    - n_top_app : 전체 앱중, 상위 n개의 앱 수 
    
    [output]
    
    """
    # 0) sampleing : df_total_app_install_sample (if necessary)
    df_app_install_sample = df_app_install.sample(n_user_sample) 
    
    # 1) generate total data
    ## app_list_sample : 모든 앱의 연관도를 계산하는 것은 비효율적이기에, 전체데이터 기준 설치 빈도 상위 n(500)개의 앱에 한하여 연관도 계산
    df_total_app_count, app_list_sample, n_total_user = gen_df_total_app_count(df_app_install = df_app_install_sample,
                                                                           n_user_sample = n_user_sample,
                                                                           n_top_app = n_top_app)

    # 2) generate target data 
    ## : 타겟 앱을 설치하고 있는 유저들의 목록(df_target_app_install) 및 숫자(n_target_user)와, 타겟 집단의 앱 설치 빈도 데이터(df_target_app_count) 생성
    df_target_app_install, df_target_app_count, n_target_user  = gen_df_target_app_count(df_app_install = df_app_install_sample,
                                                                                     n_user_sample = n_user_sample,
                                                                                     target_app = target_app)
    
    # 3) extract_n_top_app_data for get Total rate & intersect rate
    ## : total & target 데이터 설치 빈도 상위 N(500개)만을 활용하여 전체 설치 확률(total_rate) & 타겟 데이터의 설치 확률(intersect_rate)계산 산출
    df_total_app_count_sample, df_target_app_count_sample = extract_n_top_app_data(df_total_app_count = df_total_app_count, 
                                                                                   df_target_app_count = df_target_app_count,
                                                                                   n_top_app = n_top_app,
                                                                                   n_total_user = n_total_user, 
                                                                                   n_target_user = n_target_user)
    
    # 4) merge total & target (total_rate & intersect_rate)
    #       app_name 	total_rate 	intersect_rate
    # 0 	엔제리너스 카드 	0.999286 	1.000000
    # 1 	카카오톡 KakaoTalk 	0.992143 	0.997426
    # 2 	V3 Mobile Plus 2.0 	0.978714 	0.998713
    df_total_app_relation_fin = merge_total_target(df_total_app_count_sample, df_target_app_count_sample)
    
    # 5) 최종 relation 산출
    ## how : ABS, REL, GEO, weight 4가지중 하나 선택
    ## rel_weight
    ##   1. ABS_diff
    ##   2. Rel_diff
    ##   3. Geo_mean
    ##   4. weighted_Geo_mean
    df_total_app_relation_fin = calc_relataion(df_total_app_relation_fin, how, rel_weight = 0.7)

    return df_total_app_relation_fin
    
```
